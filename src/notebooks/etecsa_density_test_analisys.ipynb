```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
categories = ['baja', 'media', 'alta']
geospatial_mean_distance = [4.110, 3.793, 3.499]
geospatial_std = [0.5, 0.3, 0.4]         # Valores de desviación estándar (ejemplo)
geospatial_median_distance = [4.0, 3.75, 3.45]  # Valores de la mediana (ejemplo)

# Posiciones en el eje x
x = np.arange(len(categories))

# Creación del gráfico de barras con barras de error para la desviación estándar
plt.figure(figsize=(6, 4))
bars = plt.bar(x, geospatial_mean_distance, color=['blue', 'orange', 'green'], 
               alpha=0.7, yerr=geospatial_std, capsize=5, label='Media')

# Agregar la mediana como puntos
plt.scatter(x, geospatial_median_distance, color='red', zorder=5, label='Mediana', s=100)

# Configurar etiquetas y eje x
plt.xticks(x, categories)
plt.xlabel('Categoría', fontsize=12)
plt.ylabel('Distancia Geoespacial (km)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
