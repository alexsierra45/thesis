```python
import json
import folium
```

```python
habana_location = [23.0826, -82.2845]
```

```python
fd = open('./transport_zones.json', 'r')
transp_zones_json = json.load(fd)
fd.close()

fd = open('./voronoi_properties.json', 'r')
voronois_json = json.load(fd)
fd.close()

voronois_habana_features = list(filter(lambda f : f['properties']['province'] == 'La Habana', voronois_json['features']))
```

```python
map = folium.Map(location=habana_location, zoom_start=11, tiles='openstreetmap')
zones_geojson = folium.GeoJson(transp_zones_json, 
                               style_function = lambda x: {'weight': 1.5, 'fillOpacity' : .01}, 
                               tooltip=folium.features.GeoJsonTooltip(fields = ['NO_DE_ZONA']))

voronoi_geojson = folium.GeoJson({ 'type' : 'FeatureCollection', 'features' : voronois_habana_features },
                                style_function = lambda x: {'weight': 1.5, 'fillOpacity' : .01, 'color' : 'red'})

# voronoi_geojson.add_to(map)
zones_geojson.add_to(map)
map
```
